# Extended Kalman Filter Project
## Dependencies
Additional to the dependencies stated in the README.md file the boost library (http://www.boost.org/) is required

## Code architecture
The code was implemented as required in the following files:
- main.cpp: Main routine with WebSocket connection to simulator. Modifications were done to write all relevant data to log files for analysis in a Jupyter notebook
- measurement_package.h: Measurement definition as received by the WebSocket
- FusionEKF.h/FusionEKF.cpp: Sensor fusion routine to merge radar and laser measurements with an extended Kalman filter. For flexibility a json configuration (./data/config.json) option was included
- kalman_filter.h/kalman_filter.cpp: Implementation of the extended Kalman filter, includes the Predict() as well as Update() for Laser and UpdateEKF() for Radar measurements
- tools.h/tools.cpp: Helper functions to convert polar <-> cartesian coordinates and to calculate RMSE

## Implementation
The implementation of the Kalman filter was straight forward, since most of the codebase was already supplied by the udacity lectures. I found the major challenge to be the correct handling of the angle for radar measurements. These had to be correctly converted to the right range in order to get meaningful results. The angle conversion is handled by Tools::WrapAngle(). This helper function converts an angle to the range within min / max and is used in KalmanFilter::UpdateEKF() to preprocess the internal state x within the range of the measurement z.

## Analysis
To analyze the performance of the Kalman filter the code writes .csv files of all relevant data:
- time: current timestamp of measurement
- gt_x,gt_y,gt_vx,gt_vy: ground truth of position and velocity
- est_x,est_y,est_vx,est_vy: estimate for position and velocity
- rmse_x,rmse_y,rmse_vx,rmse_vy: RMSE between ground truth and velocty

This data is plotted by the Jupyter notebook with create_plots() to analyze the performance for various configuration parameters.

```python
import matplotlib.pyplot as plt
import csv
import json
from pprint import pprint
import os

%matplotlib inline

MAX_RMSE_POS = 0.11
MAX_RMSE_VEL = 0.52

def create_plots(dataset):
    print('CONFIGURATION:')
    with open('./data/' + dataset + '.json') as json_data:
        d = json.load(json_data)
        json_data.close()
        pprint(d)
    
    datafiles = ['./data/' + dataset + '_dataset' + str(i) + '.dat' for i in [1,2]]
    datafiles = [d for d in datafiles if os.path.exists(d)]

    for df in datafiles:
        print('DATAFILE: ' + df)
        with open(df, 'r') as csvfile:
            csvreader = csv.reader(csvfile, delimiter=',', quotechar='|')

            header_row = next(csvreader)

            data = dict.fromkeys(header_row)
            for key in data:
                data[key] = []

            for row in csvreader:
                for i,item in enumerate(row):
                    data[header_row[i]].append(float(item))

        time_min = data['time'][0]
        time_max = data['time'][-1]
        time_bounds = [time_min, time_max]

        plt.figure(figsize=(20,10))

        plt.subplot(2,2,1)
        plt.plot(data['gt_x'], data['gt_y'], 'b-', data['est_x'], data['est_y'], 'r-')
        plt.xlabel('x pos')
        plt.ylabel('y pos')
        plt.legend(['ground truth', 'estimate'])

        plt.subplot(2,2,2)
        plt.plot(data['gt_vx'], data['gt_vy'], 'b-', data['est_vx'], data['est_vy'], 'r-')
        plt.xlabel('x vel')
        plt.ylabel('y vel')
        plt.legend(['ground truth', 'estimate'])

        plt.subplot(2,2,3)
        plt.plot(data['time'], data['rmse_x'], 'b-', data['time'], data['rmse_y'], 'r-',)
        plt.plot(time_bounds, [MAX_RMSE_POS, MAX_RMSE_POS], 'g--')
        plt.ylabel('rmse pos')
        plt.ylim([0,0.25])
        plt.legend(['x', 'y', 'limit'])

        plt.subplot(2,2,4)
        plt.plot(data['time'], data['rmse_vx'], 'b-', data['time'], data['rmse_vy'], 'r-',)
        plt.plot(time_bounds, [MAX_RMSE_VEL, MAX_RMSE_VEL], 'g--')
        plt.ylabel('rms vel')
        plt.ylim([0,2])
        plt.legend(['x', 'y', 'limit'])

        plt.show()
```

# Default settings
To start the default settings from the provided codebase are used. These already give quite a satisfactory result with the RMSE of position / velocity meeting the target values of 0.11 / 0.52 (green dashed line) with both dataset1 and dataset2 of the simulator.

```python
dataset = 'config01_default'

create_plots(dataset)
```

# Radar / Laser measurements
To analyze the influence of Laser and Radar measurements one was turned off and the parameters for the other were tune (more / less measurement noise)

## Laser only
With laser only the overall accuracy drops and the measurement noise does not have much influence. The overall performance increase with 

```python
datasets = ['config02_laser', 'config03_laser_morenoise', 'config04_laser_lessnoise']

for d in datasets:
    create_plots(d)
```

# Radar only
With Radar measurements only the position data is quite off, as can be expected since the Radar provides more accurate velocity information. The overall performance is slightly better with less measurement noise.

```python
datasets = ['config05_radar', 'config06_radar_morenoise', 'config07_radar_lessnoise']

for d in datasets:
    create_plots(d)
```

# Process noise
As last step the process noise was varied with the default configuration. Reducing the noise results in slightly worse performance, while increasing the process noise improves the performance slightly.

```python
datasets = ['config08_lessprocessnoise', 'config09_moreprocessnoise']

for d in datasets:
    create_plots(d)
```

# Discussion
The extended Kalman filter performs quite well in fusing the laser and radar measurements. The laser measurements seem to have a larger impact on the accuracy of the filter output. The influence of process and measurement noise is smaller than initially expected.

## Possible improvements
The Tools::WrapAngle() functions is quite insufficient (non deterministic) and can be improved by using fnmod(). Further an optimization by merging the identical parts of KalmanFilter::Update() and KalmanFilter::UpdateEKF() is possible.
